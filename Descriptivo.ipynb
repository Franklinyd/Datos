# **Estadistica descriptiva**

```python
tc=pd.crosstab(Data['CL1'], Data['CL9'],normalize='index',margins=True)
tc=tc*100
tc=tc.round(1)
tc
```

```python
import seaborn as sns
sns.heatmap(tc, annot=True, cmap="YlGnBu")
plt.title('CL1 Vs CL9')
plt.show()
```

```python
plt.figure(figsize=(10,5))
bar_chart = plt.bar(tc.index, tc.values.ravel(), width = 0.4,color='blue',edgecolor='black', linewidth=1.5)
plt.xticks(rotation=45)
plt.xlabel('columna_A')
plt.ylabel('Frecuencia')
plt.title('Tabla de contingencia')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear un ejemplo de DataFrame
df = pd.DataFrame({'A': np.random.randint(0,2,size=100),
                  'B': np.random.randint(0,3,size=100)})

# Crear la tabla de contingencia
ctab = pd.crosstab(df['A'], df['B'])

# Crear el gráfico de barras apiladas
fig, ax = plt.subplots(figsize=(10,5))

barWidth = 0.4
bar_A = np.arange(len(ctab.iloc[:,0]))
bar_B = [x + barWidth for x in bar_A]
bar_C = [x + barWidth for x in bar_B]

bar_A = ax.bar(bar_A, ctab.iloc[:,0], width = barWidth,color='blue',edgecolor='black', linewidth=1.5)
bar_B = ax.bar(bar_B, ctab.iloc[:,1], width = barWidth,color='green',edgecolor='black', linewidth=1.5)
bar_C = ax.bar(bar_C, ctab.iloc[:,2], width = barWidth,color='red',edgecolor='black', linewidth=1.5)

ax.set_xticks([r + barWidth for r in range(len(ctab.iloc[:,0]))])
ax.set_xticklabels(ctab.index)
ax.set_xlabel('columna_A')
ax.set_ylabel('Frecuencia')
ax.set_title('Tabla de contingencia')

plt.legend((bar_A,bar_B,bar_C), ('Valor 0', 'Valor 1', 'Valor 2'))
```
